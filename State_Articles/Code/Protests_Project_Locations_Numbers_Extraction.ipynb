```python
from geotext import GeoText

def geo_places(text):
    places = GeoText(str(text))
    cities = list(places.cities)
    return(cities)

def geo_country_mentions(text):
    countries = GeoText(str(text)).country_mentions
    country_list = list(countries)
    return(country_list)
```

```python
def state_location_mentions(state_news_df):
    state_city_list = []
    state_country_list = []

    for i in range(state_news_df.shape[0]):
        if len(state_news_df['city_mentions']) > 0:
            state_city_list = state_city_list + [e for e in state_news_df.loc[i,'city_mentions']]
        if len(state_news_df['country_mentions']) > 0:
            state_country_list = state_country_list + [e for e in state_news_df.loc[i,'country_mentions']]
        
    return(state_city_list, state_country_list)
```

```python
# Extract all numbers - from title and article content
def numbers(text):
    res = [int(i) for i in text.split() if i.isdigit()]
    return(res)
```

```python
import re

# Extract the # of arrest(s), protester(s) - from title and article content

def method_1_arrest(text):
    candidate_arrest = re.search(r'(\w+\s+){0,3}arrest(\w+\s+){0,3}', text, re.IGNORECASE)
    try: 
        output = candidate_arrest.group(0)
        return(output)
    except:
        return([])  

def method_2_arrest(text):
    candidate_arrest = r"(?i)(?:\barrest\D{0,20})([0-9][0-9,]*)[^.,]|([0-9][0-9,]*)[^.,](?:\D{0,20}arrest)"
    output_list = re.findall(candidate_arrest,text)
    final_list = [item[0] if item[0] != '' else item[1] for item in output_list]
    return(final_list)

def method_3_arrest(text):
    candidate_arrest = r"(?:(?<!\S)(\d+(?:,\d+)?)(?!\S)\D*\barrest|\barrestD*(?<!\S)(\d+(?:,\d+)?)(?!\S))"
    output_list = re.findall(candidate_arrest,text)
    final_list = [item[0] if item[0] != '' else item[1] for item in output_list]
    return(final_list)

def method_1_arrests(text):
    candidate_arrest = re.search(r'(\w+\s+){0,3}arrests(\w+\s+){0,3}', text, re.IGNORECASE)
    try: 
        output = candidate_arrest.group(0)
        return(output)
    except:
        return([])  

def method_2_arrests(text):
    candidate_arrest = r"(?i)(?:\barrests\D{0,20})([0-9][0-9,]*)[^.,]|([0-9][0-9,]*)[^.,](?:\D{0,20}arrests)"
    output_list = re.findall(candidate_arrest,text)
    final_list = [item[0] if item[0] != '' else item[1] for item in output_list]
    return(final_list)

def method_3_arrests(text):
    candidate_arrest = r"(?:(?<!\S)(\d+(?:,\d+)?)(?!\S)\D*\barrests|\barrestsD*(?<!\S)(\d+(?:,\d+)?)(?!\S))"
    output_list = re.findall(candidate_arrest,text)
    final_list = [item[0] if item[0] != '' else item[1] for item in output_list]
    return(final_list)

def method_1_protester(text):
    candidate_arrest = re.search(r'(\w+\s+){0,3}protester(\w+\s+){0,3}', text, re.IGNORECASE)
    try: 
        output = candidate_arrest.group(0)
        return(output)
    except:
        return([])  

def method_2_protester(text):
    candidate_arrest = r"(?i)(?:\bprotester\D{0,20})([0-9][0-9,]*)[^.,]|([0-9][0-9,]*)[^.,](?:\D{0,20}protester)"
    output_list = re.findall(candidate_arrest,text)
    final_list = [item[0] if item[0] != '' else item[1] for item in output_list]
    return(final_list)

def method_3_protester(text):
    candidate_arrest = r"(?:(?<!\S)(\d+(?:,\d+)?)(?!\S)\D*\bprotester|\bprotesterD*(?<!\S)(\d+(?:,\d+)?)(?!\S))"
    output_list = re.findall(candidate_arrest,text)
    final_list = [item[0] if item[0] != '' else item[1] for item in output_list]
    return(final_list)

def method_1_protesters(text):
    candidate_arrest = re.search(r'(\w+\s+){0,3}protesters(\w+\s+){0,3}', text, re.IGNORECASE)
    try: 
        output = candidate_arrest.group(0)
        return(output)
    except:
        return([])  

def method_2_protesters(text):
    candidate_arrest = r"(?i)(?:\bprotesters\D{0,20})([0-9][0-9,]*)[^.,]|([0-9][0-9,]*)[^.,](?:\D{0,20}protesters)"
    output_list = re.findall(candidate_arrest,text)
    final_list = [item[0] if item[0] != '' else item[1] for item in output_list]
    return(final_list)

def method_3_protesters(text):
    candidate_arrest = r"(?:(?<!\S)(\d+(?:,\d+)?)(?!\S)\D*\bprotesters|\bprotestersD*(?<!\S)(\d+(?:,\d+)?)(?!\S))"
    output_list = re.findall(candidate_arrest,text)
    final_list = [item[0] if item[0] != '' else item[1] for item in output_list]
    return(final_list)
```
